<a name="data-preprocessing-stage-two"></a>
### 3A. Preprocessing - Stage Two

```python
# Function to extract Mel Spectrograms and MFCCs to use in the models and store in JSON file
def preprocess_dataset(dataset_path, json_path, num_samples, num_mfcc=13, n_fft=2048, hop_length=512):

    # Dictionary to temporarily store mapping, labels, MFCCs, spectrograms and filenames
    data = {
        'mapping': [],
        'labels': [],
        'MFCCs': [],
        'mel_specs': [],
        'files': []
    }

    # Loop through all sub directories
    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):

        # Ensure we're at sub-folder level
        if dirpath is not dataset_path:

            # Save label in the mapping
            label = dirpath.split('/')[-1]
            data['mapping'].append(label)
            print("\nProcessing: '{}'".format(label))

            # Process all audio files in sub directory and store MFCCs
            for f in filenames:
                file_path = os.path.join(dirpath, f)

                # Load audio file and slice it to ensure length consistency among different files
                signal, sample_rate = librosa.load(file_path)

                # Drop audio files with less than pre-decided number of samples
                if len(signal) >= num_samples:

                    # Ensure consistency of the length of the signal
                    signal = signal[:num_samples]

                    # Extract MFCCs
                    MFCCs = librosa.feature.mfcc(signal, 
                                                 sample_rate, 
                                                 n_mfcc=num_mfcc, 
                                                 n_fft=n_fft,
                                                 hop_length=hop_length)

                    # Extract Mel Spectrograms
                    mel_specs = librosa.feature.melspectrogram(signal,
                                                               sample_rate, 
                                                               n_fft=n_fft,
                                                               hop_length=hop_length)

                    # Append data in dictionary
                    data['MFCCs'].append(MFCCs.T.tolist())
                    data['mel_specs'].append(mel_specs.T.tolist())
                    data['labels'].append(i-1)
                    data['files'].append(file_path)
                    print("{}: {}".format(file_path, i-1))

    # Save data in JSON file for re-using later
    with open(json_path, 'w') as file_path:
        json.dump(data, file_path, indent=5)
```

```python
us_dataset_path = 'data/ultrasuite_top40'
us_json_path = 'ultrasuite_data.json'
num_samples = 11025
```

```python
preprocess_dataset(us_dataset_path, us_json_path, num_samples)
```

```python
# Function to load the data from the JSON file depending on selected feature
def load_data(data_path, feature):

    with open(data_path, 'r') as file_path:
        data = json.load(file_path)

    X = np.array(data[feature])
    y = np.array(data['labels'])

    print('Datasets loaded...')
    
    return X, y
```

```python
# Function to create training, test and validation datasets
def create_train_test(data_path, feature, test_size=0.2, val_size=0.2):

    # Load dataset
    X, y = load_data(data_path, feature)

    # Create train, test and validation splits
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=val_size)

    # Increase the dimension of the NumPy array for each split - WHY?
    X_train = X_train[..., np.newaxis]
    X_test = X_test[..., np.newaxis]
    X_val = X_val[..., np.newaxis]

    return X_train, y_train, X_val, y_val, X_test, y_test
```

```python
# Function for visualising results
def visualise_results(results):
    history = results.history

    plt.figure(figsize=(20,8))
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    plt.subplot(1, 2, 1)
    plt.plot(history['val_loss'])
    plt.plot(history['loss'])
    plt.legend(['Validation Loss', 'Training Loss'], fontsize=12)
    plt.title('Loss', fontsize=18)
    plt.xlabel('Epochs', fontsize=14)
    plt.ylabel('Loss', fontsize=14)
    
    plt.subplot(1, 2, 2)
    plt.plot(history['val_acc'])
    plt.plot(history['acc'])
    plt.legend(['Validation Accuracy', 'Training Accuracy'], fontsize=12)
    plt.title('Accuracy', fontsize=18)
    plt.xlabel('Epochs', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.show()
```

```python
# Create the train, test and validation datasets for the Ultrasuite Top 40 dataset
us_data_path = 'ultrasuite_data.json'
X_train, y_train, X_val, y_val, X_test, y_test = create_train_test(us_data_path, 'MFCCs')
```

```python
# Explore the dataset again, checking array shapes
print ("Number of training samples: " + str(X_train.shape[0]))
print ("Number of testing samples: " + str(X_test.shape[0]))
print ("Number of validation samples: " + str(X_val.shape[0]))
print ("X_train shape: " + str(X_train.shape))
print ("y_train shape: " + str(y_train.shape))
print ("X_test shape: " + str(X_test.shape))
print ("y_test shape: " + str(y_test.shape))
print ("X_val shape: " + str(X_val.shape))
print ("y_val shape: " + str(y_val.shape))
```

```python
y_test[:10]
```

```python
def reformat_y(y):
    y = LabelEncoder().fit_transform(y)
    y = tf.keras.utils.to_categorical(y)
    return y
```

```python
train_y = reformat_y(y_train)
test_y = reformat_y(y_test)
val_y = reformat_y(y_val)
```

```python
test_y[:10]
```

<a name="baseline-model"></a>
### 3B. Create a simple baseline model

```python
# Build a baseline model
def build_baseline_model(input_shape, loss_func='categorical_crossentropy', learning_rate=0.0001):
    
    baseline_model = tf.keras.models.Sequential()
    baseline_model.add(tf.keras.layers.InputLayer(input_shape=input_shape))
    baseline_model.add(tf.keras.layers.Flatten())
    baseline_model.add(tf.keras.layers.BatchNormalization())
    baseline_model.add(tf.keras.layers.Dense(40, activation='softmax'))
    
    # Set optimizer and learning rate
    optimiser = tf.optimizers.Adam(learning_rate=learning_rate)
    
    # Compile the baseline model
    baseline_model.compile(loss=loss_func,
                           optimizer=optimiser,
                           metrics=['acc'])

    # Print summary for model
    baseline_model.summary()
    
    return baseline_model
```

```python
# Function for fitting the model
def fit_model(model, epochs, batch_size, patience, X_train, y_train, X_val, y_val):

    # Define early stopping criteria
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy', min_delta=0.001, patience=patience)

    # Fit the model
    results = model.fit(X_train,
                        y_train,
                        epochs=epochs,
                        batch_size=batch_size,
                        validation_data=(X_val, y_val),
                        callbacks=[early_stopping])
    return results
```

```python
# Create baseline network
input_shape = (X_train[0].shape)
baseline_model = build_baseline_model(input_shape, learning_rate=0.0001)

 
# Fit network
results_1 = fit_model(baseline_model, 40, 32, 5, X_train, train_y, X_val, val_y)
```

```python
# Visualise the loss and accuracy of the training and validation sets across epochs
visualise_results(results_1)
```

```python
# Evaluate the training results
results_1_train = baseline_model.evaluate(X_train, train_y)
results_1_train
```

```python
# Evaluate the test results
results_1_test = baseline_model.evaluate(X_test, test_y)
results_1_test
```

#### Conclusion

It's starting point but the training accuracy is only 75%, but the validation accuracy is slowly converging.

The model is clearly overfitting the training data and is not generalising well when shown unseen data.

```python
# Create baseline network
input_shape = (X_train[0].shape)
baseline_model_2 = build_baseline_model(input_shape, learning_rate=0.001)

 
# Fit network
results_2 = fit_model(baseline_model_2, 30, 128, 3, X_train, train_y, X_val, val_y)
```

```python
# Visualise the loss and accuracy of the training and validation sets across epochs
visualise_results(results_2)
```

```python
# Evaluate the training results
results_2_train = baseline_model_2.evaluate(X_train, train_y)
results_2_train
```

```python
# Evaluate the test results
results_2_test = baseline_model_2.evaluate(X_test, test_y)
results_2_test
```

#### Conclusion

We have improved the training accuracy of the baseline model, to almost 88%, by increasing the learning rate and reducing the number of epochs but this has only served to exascerbate the issue of overfitting with test accuracy not improving on 68%.
